```python
Data_Num = Data_R_NoD.select_dtypes(include = np.number)
Q1 =  Data_Num.quantile(0.25)
Q3 =  Data_Num.quantile(0.70)
IQR = Q3 - Q1
Upper_Bound = Q3 + 1.5*IQR
Lower_Bound = Q1 - 1.5*IQR
outliers = ((Data_Num < Lower_Bound)|(Data_Num  > Upper_Bound))
outliers.sum()
```

```python
#Now after i found how many outliers there s in each feature i need to find those outliers and their rows and their values to see if they can be considered outliers or not as this dataset is about cars so it makes sense to find some really expensive cars and really cheap cars and i don t want to loose data 
```
